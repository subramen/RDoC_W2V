```python
import gensim
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
import time
```

```python
fname = 'data/one-abstract-per-line.txt'
# sentenceG = (line.split() for line in open(fname))
class sentGen:
    def __init__(self, f, limit=None):
        self.f=f
        self.li = limit
    def __iter__(self):
        c=0
        for l in open(self.f):
            if self.li and c>=self.li:
                break
            c+=1
            yield l.split()
            
```

```python
start = time.time()
s=sentGen(fname)
model = Word2Vec(s, size=300, window=5, min_count=1, workers=4)
word_vectors = model.wv
duration = time.time()-start
print(duration)
```
